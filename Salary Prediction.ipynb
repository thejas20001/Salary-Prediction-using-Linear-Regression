## Salary Prediction

```python
import pandas as pd
```

```python
data = pd.read_csv("Salary_Data-210418-183210.csv")
```

```python
## salary as labels and experience as features
```

```python
data
```

```python
x = data['YearsExperience'].values
```

```python
y = data['Salary'].values
```

```python
x
```

```python
y
```

```python
x.ndim
```

```python
y.ndim
```

```python
x = x.reshape(-1,1)
```

```python
y = y.reshape(-1,1)
```

```python
x
```

```python
y
```

```python
x.ndim
```

```python
y.ndim
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
```

```python
len(x)
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.20)
```

```python
xtrain
```

```python
xtest
```

```python
len(xtest)
```

```python
len(xtrain)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
abc = LinearRegression(0)
```

### Training Model

```python
abc.fit(xtrain,ytrain)
```

```python
abc.intercept_
```

```python
abc.coef_
```

```python
xpredict = abc.predict(xtrain)
```

```python
xpredict
```

```python
plt.scatter(xtrain,ytrain,color='green')
plt.plot(xtrain,xpredict,'r')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

## Testing of the model

```python
xpredicttest = abc.predict(xtest)
```

```python
xpredicttest
```

```python
plt.scatter(xtrain,ytrain,color = 'green')
plt.scatter(xtest,ytest,color = 'blue')
plt.plot(xtest,xpredicttest,color = 'red')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```
